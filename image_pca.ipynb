```python
import os
import cv2
import numpy as np
```

```python
data_dir = '/home/aicenter/Documents/hsu/data/test_dataset/tw_labels/datasets/image/'
base_image_paths = [data_dir + filename for filename in os.listdir(data_dir)]
base_image_set = []
for image_path in base_image_paths:
    image = cv2.imread(image_path)
    image = cv2.resize(image, (400, 400))
    image_Lab = cv2.cvtColor(image, cv2.COLOR_RGB2Lab)
    image = image_Lab[:, :, 0]
    base_image_set.append(image)
base_image_set = np.array(base_image_set)
base_image_set = base_image_set.reshape(len(base_image_paths), -1)
base_image_mean = np.mean(base_image_set, axis=0)
base_eigenvectors, base_eigenvalues, base_V = np.linalg.svd((base_image_set-base_image_mean).T, full_matrices=False)
```

```python
index = 0
cv2.imwrite('base_original_crack.png', (base_image_set[index]).reshape(400, 400))
pca = base_eigenvectors.T[:20]
recons_image = ((base_image_set[index] - base_image_mean)@pca.T)@pca + base_image_mean
cv2.imwrite('base_pca_crack.png', (recons_image).reshape(400, 400))
```

```python
data_dir = '/home/aicenter/Documents/hsu/data/public_dataset/crack_segmentation_dataset/groupby/DeepCrack/'
image_paths = [data_dir + filename for filename in os.listdir(data_dir)]
image_set = []
for image_path in image_paths:
    image = cv2.imread(image_path)
    image = cv2.resize(image, (400, 400))
    image_Lab = cv2.cvtColor(image, cv2.COLOR_RGB2Lab)
    image = image_Lab[:, :, 0]
    image_set.append(image)
image_set = np.array(image_set)
image_set = image_set.reshape(len(image_paths), -1)
image_mean = np.mean(image_set, axis=0)
eigenvectors, eigenvalues, V = np.linalg.svd((image_set-image_mean).T, full_matrices=False)
```

```python
index = 0
print(image_paths[index])
cv2.imwrite('original_crack.png', (image_set[index]).reshape(400, 400))
pca = eigenvectors.T[:20]
recons_image = ((base_image_set[index] - base_image_mean)@pca.T)@pca + base_image_mean
cv2.imwrite('pca_crack.png', (recons_image).reshape(400, 400))
```
